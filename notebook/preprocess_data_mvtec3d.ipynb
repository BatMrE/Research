```python
import os
import shutil

root_data = r"D:\MS\Research\Thesis\Code\github_MVtec3D\Data\full_data\processed_data"
output_dir = r"D:\MS\Research\Thesis\Code\github_MVtec3D\Data\full_data\processed_anomaly_data"

subfolders = ["gt", "rgb", "xyz"]

# Create output directories
for sf in subfolders:
    os.makedirs(os.path.join(output_dir, sf), exist_ok=True)

# Traverse each category (bagel, cable_gland, carrot, ...)
for category in os.listdir(root_data):
    category_path = os.path.join(root_data, category)
    if not os.path.isdir(category_path):
        continue

    test_path = os.path.join(category_path, "test")
    if not os.path.isdir(test_path):
        continue

    # Iterate through each defect type (combined, contamination, crack, etc.)
    for defect_type in os.listdir(test_path):
        defect_path = os.path.join(test_path, defect_type)
        if not os.path.isdir(defect_path):
            continue

        for sf in subfolders:
            sf_path = os.path.join(defect_path, sf)
            if not os.path.exists(sf_path):
                continue

            # Copy each file with new name
            for filename in sorted(os.listdir(sf_path)):
                src_file = os.path.join(sf_path, filename)
                if not os.path.isfile(src_file):
                    continue

                new_filename = f"{category}_{filename}"
                dest_file = os.path.join(output_dir, sf, new_filename)

                shutil.copy(src_file, dest_file)
```

```python
import os
import shutil

from pathlib import Path

# Input and output paths
root_dir = Path(r"D:\MS\Research\Thesis\Code\github_MVtec3D\Data\full_data")
collection_dir = Path(r"D:\MS\Research\Thesis\Code\github_MVtec3D\train_data\collection")
(gt_out, rgb_out, xyz_out) = (collection_dir / "gt", collection_dir / "rgb", collection_dir / "xyz")

# Create output directories
gt_out.mkdir(parents=True, exist_ok=True)
rgb_out.mkdir(parents=True, exist_ok=True)
xyz_out.mkdir(parents=True, exist_ok=True)

# Traverse all "test" folders recursively
for test_dir in root_dir.rglob("test"):
    if not test_dir.is_dir():
        continue

    for sub in test_dir.iterdir():
        if sub.name == "good" or not sub.is_dir():
            continue

        parent_tag = f"{test_dir.parent.name}_{sub.name}"

        # Copy files from gt, rgb, xyz folders
        for category in ['gt', 'rgb', 'xyz']:
            src_dir = sub / category
            dst_dir = {"gt": gt_out, "rgb": rgb_out, "xyz": xyz_out}[category]

            if not src_dir.exists():
                continue

            for file in src_dir.iterdir():
                if file.is_file():
                    new_name = f"{parent_tag}_{file.name}"
                    shutil.copy(file, dst_dir / new_name)

print("Collection complete.")
```

```python
import os
import shutil

# List of shape folders you want to combine
shape_folders = ['bagel', 'cable_gland', 'carrot', 'cookie', 'dowel', 'foam', 'peach', 'potato', 'rope', 'tire']  # Add more if needed

# Base path
base_path = r'D:\MS\Research\Thesis\Code\github_MVtec3D\Data\full_data\mv_3d_anomaly'

# Output folder where combined files will go
output_base = r'D:\MS\Research\Thesis\Code\github_MVtec3D\Data\full_data\processed_good_data'
output_rgb = os.path.join(output_base, 'rgb')
output_xyz = os.path.join(output_base, 'xyz')

# Create output folders if not exist
os.makedirs(output_rgb, exist_ok=True)
os.makedirs(output_xyz, exist_ok=True)

# Copy and rename files
for shape in shape_folders:
    shape_path = os.path.join(base_path, shape, 'train', 'good')
    for modality in ['rgb', 'xyz']:
        input_folder = os.path.join(shape_path, modality)
        output_folder = output_rgb if modality == 'rgb' else output_xyz

        if not os.path.isdir(input_folder):
            print(f"Missing folder: {input_folder}")
            continue

        for file_name in os.listdir(input_folder):
            src_file = os.path.join(input_folder, file_name)
            new_name = f"{shape}_{file_name}"
            dst_file = os.path.join(output_folder, new_name)
            shutil.copy(src_file, dst_file)

print("✅ Done combining and copying files.")
```
